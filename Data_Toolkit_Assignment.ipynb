2. How does broadcasting work in NumPy?

Broadcasting in NumPy allows us to perform arithmetic operations on arrays of different shapes without reshaping them. It automatically adjusts the smaller array to match the larger array's shape by replicating its values along the necessary dimensions. This makes element-wise operations more efficient by reducing memory usage and eliminating the need for loops.

For Example:

import numpy as np

array_2d = np.array([[1, 2, 3], [4, 5, 6]])  
scalar = 10  

result = array_2d + scalar
print(result)

Output: [[11 12 13]
        [14 15 16]]

Working of Broadcasting in NumPy:

Broadcasting applies specific rules to find whether two arrays can be aligned for operations or not that are:

1. Check Dimensions: Ensure the arrays have the same number of dimensions or expandable dimensions.
2. Dimension Padding: If arrays have different numbers of dimensions the smaller array is left-padded with ones.
3. Shape Compatibility: Two dimensions are compatible if they are equal or one of them is 1.

6. What are the differences between NumPy arrays and Python lists?


Difference between numpy arrays and python lists are as follow:

Python Lists:

1. Element Overhead: Lists in Python store additional information about each element, such as its type and reference count. This overhead can be significant when dealing with a large number of elements.
2. Datatype: Lists can hold different data types, but this can decrease memory efficiency and slow numerical operations.
3. Memory Fragmentation: Lists may not store elements in contiguous memory locations, causing memory fragmentation and inefficiency.
4. Performance: Lists are not optimized for numerical computations and may have slower mathematical operations due to Python's interpretation overhead. They are generally used as general-purpose data structures.
5. Functionality: Lists can store any data type, but lack specialized NumPy functions for numerical operations.

Numpy Arrays

1. Homogeneous Data: NumPy arrays store elements of the same data type, making them more compact and memory-efficient than lists.
2. Fixed Data Type: NumPy arrays have a fixed data type, reducing memory overhead by eliminating the need to store type information for each element.
3. Contiguous Memory: NumPy arrays store elements in adjacent memory locations, reducing fragmentation and allowing for efficient access.
4. Array Metadata: NumPy arrays have extra metadata like shape, strides, and data type. However, this overhead is usually smaller than the per-element overhead in lists.
5. Performance: NumPy arrays are optimized for numerical computations, with efficient element-wise operations and mathematical functions. These operations are implemented in C, resulting in faster performance than equivalent operations on lists.

15. How does NumPy handle multidimensional arrays?

NumPy's core functionality revolves around the ndarray object, which efficiently handles multidimensional arrays. Here's how:

1. Homogeneous Data:

(A). NumPy arrays store elements of the same data type, unlike Python lists that can hold mixed types. This uniformity enables optimized memory usage and faster computations.

2. Memory Layout:

(A). Multidimensional arrays are stored as a contiguous block of memory, treated as a 1D array internally. The shape and strides information define how to interpret this 1D block as a multidimensional structure.

(B). This contiguous storage allows for efficient access and manipulation of array elements, leveraging the principle of locality of reference.

3. Shape and Axes:

(A). The shape attribute defines the dimensions of the array. For instance, a shape of (2, 3) represents a 2D array with 2 rows and 3 columns.

(B). Dimensions are referred to as axes. In a 2D array, axis 0 represents rows, and axis 1 represents columns.

4. Indexing and Slicing:

(A). NumPy allows powerful indexing and slicing using tuples of integers or slice objects.

(B).Slicing enables extracting subarrays with specific ranges along each axis.

5. Broadcasting:

(A). NumPy supports broadcasting, which allows element-wise operations between arrays of different shapes under certain conditions. This feature avoids explicit looping and simplifies code.

6. Reshaping and Views:

(A).The reshape() function can alter the shape of an array without changing its underlying data.

(B). Slicing and other operations often return views of the original array, sharing the same data buffer, which avoids unnecessary copying.

7. Mathematical Functions:

(A). NumPy provides a wide range of mathematical and statistical functions that operate efficiently on multidimensional arrays.

17. Explain the difference between apply() and map() in Pandas?

apply() and map() are both powerful functions in Pandas used for applying operations to data, but they differ in their scope and usage:

1. map()

(A). Scope: Primarily for element-wise operations on Pandas Series.

(B). Functionality: It substitutes each value in a Series with another value, based on a mapping defined by a dictionary, a Series, or a function.

(C). Efficiency: Generally faster for simple element-wise transformations on Series.

(D). Use Case: Ideal for value replacement, like mapping codes to labels, and simple transformations.

(E). Input: Accepts a dictionary, Series, or a callable function.

2. apply()

(A). Scope: Operates on both Pandas Series and DataFrames.

(B). Functionality: Applies a function along an axis of a DataFrame or to values of a Series. It can be used for more complex operations and aggregations.

(C). Flexibility: Highly flexible, allowing functions to be applied along rows or columns of a DataFrame.

(D). Use Case: Suitable for complex transformations, aggregations, and when operations need to consider multiple rows or columns.

(E). Input: Accepts callable functions only.

(F). Performance: Can be slower than map() for simple element-wise operations due to function call overhead.

4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot

#Example DataFrame
data = {
    'A': [1, 2, 3, 4, 5],
    'B': [5, 4, 3, 2, 1],
    'C': [10, 20, 30, 40, 50]
}
df = pd.DataFrame(data)

#Calculate Correlation Matrix
corr_matrix = df.corr()

# Visualize with Seaborn Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidth=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```
